2. Carga de los estados de la Republica Mexicana en un GeoDataFrame

```python
estados_mex = geopandas.read_file("/content/gdrive/MyDrive/Colab Notebooks/mexican states/mexican-states.shp")
estados_mex
```

3. Carga de los datos de la localizacion de sucursales Costco

```python
df = geopandas.read_file("/content/gdrive/MyDrive/Colab Notebooks/tiendaCostco.csv")
df
```

4. Convertir datos de localizacion en un GeoDataFrame creando puntos

```python
sucursales_costco = geopandas.GeoDataFrame(df, geometry= geopandas.points_from_xy(df.longitud, df.latitud))
sucursales_costco = sucursales_costco.set_crs(estados_mex.crs)
sucursales_costco
```

5. Agregar columna de region a donde pertence el estado, filtrado del dataframe y separacion por regiones

```python
estados_mex["region"] = ["centro oeste","centro oeste","centro","centro","centro",
"centro","centro","centro","noreste","noreste",
"noreste","noroeste","centro oeste","centro oeste","centro oeste",
"centro oeste","centro oeste","noreste","noreste","sureste",
"sureste","noroeste","sureste","centro oeste","sureste",
"sureste","sureste","noroeste","noroeste","centro oeste","sureste","sureste"]
estados_mex = estados_mex.loc[:, ["ISO3166-2", "name", "region", "geometry"]]

regiones = estados_mex.dissolve(by='region')
regiones.plot(cmap= "rainbow")
```

6. Asegura que ambos GeoDataFrames utilicen el CRS epsg=3395 que usa metros como medida de distancia.

```python
regiones = regiones.to_crs(epsg= 3395)
sucursales_costco = sucursales_costco.to_crs(epsg= 3395)
```

7. Crea un buffer de 120 km alrededor de los puntos de localización de cada sucursal.

```python
sucursales_costco["geometry"] = sucursales_costco.buffer(120000)
sucursales_costco.plot()
```

8. Grafica la intersección de las regiones con los círculos creados a partir de la localización de las sucursales.

```python
interseccion = geopandas.overlay(regiones, sucursales_costco, how= "intersection")
interseccion.plot(alpha= 0.5, cmap= "rainbow", edgecolor= "black")
```

9. Identifica las regiones en donde se requiere más cobertura e indica en qué regiones del país NO es conveniente abrir una nueva sucursal debido a la falta de cobertura de entregas a domicilio.

```python
#obtener el mapa de mexico para obtener contorno del mapa
mex = estados_mex
mex = mex.dissolve()
mex = mex.to_crs(epsg= 3395)
#Crear graficas
fig, ax= plt.subplots()
ax.set_title("Regiones que requieren mayor cobertura")
#graficar mapa de mexico
mapa = mex.boundary.plot(ax= ax, color = "black")
#graficar interseccion
inter = geopandas.overlay(regiones, sucursales_costco, how= "intersection")
inter.plot(ax= ax, cmap= "rainbow")
#Crear graficas
fig, ax= plt.subplots()
ax.set_title("Regiones donde NO es conveniente abrir sucursales")
#graficar mapa de mexico
mapa = mex.boundary.plot(ax= ax, color = "black")
#graficar diferencia
diff = geopandas.overlay(regiones, sucursales_costco, how= "difference")
diff.plot(ax= ax, cmap= "rainbow")
```

Decision: Expandir las sucursales en las regiones que requieren mayor cobertura de acuerdo con el mapa: Region Centro y Region Centro Oeste

No expandir en regiones: Noreste y Noroeste
